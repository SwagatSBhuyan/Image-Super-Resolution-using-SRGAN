```python
import numpy as np 
import matplotlib.pyplot as plt 
  
X = ['BiCubic','SRCNN','SRResNet','SRGAN']
Set5 = [0.811, 0.886, 0.8788, 0.7903]
Set14 = [0.7163, 0.771, 0.7101, 0.6826]
BSD100 = [0.6681, 0.742, 0.619, 0.67] 
  
X_axis = np.arange(len(X))

plt.figure().set_figwidth(10)
plt.figure().set_figwidth(12)  

plt.bar(X_axis - 0.2, Set5, 0.3, label = 'Set5')
plt.bar(X_axis + 0.0, Set14, 0.3, label = 'Set14')
plt.bar(X_axis + 0.2, BSD100, 0.3, label = 'BSD100')
  


plt.xticks(X_axis, X)
plt.xlabel("SR models")
plt.ylabel("SSIM")
plt.title("SSIM scores for different SR models")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# creating the dataset
data = {
    'BiCubic': 1.12, 
    'SRCNN': 2.33, 
    'SRResNet': 2.98,
		'SRGAN': 4.11,
    'HR (original)': 4.96
}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(courses, values, color ='orange',	width = 0.4)

plt.xlabel("SR Model (compared to HR)")
plt.ylabel("MOS")
plt.title("Comparison of models based on Mean Opinion Scores (Average)")
plt.show()
```
